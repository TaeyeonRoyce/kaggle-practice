```python
# 라이브러리 import
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD

import numpy as np
```

```python
# 데이터 구축
# 학습 데이터는 (15, 3), 결과 데이터는 (15,)
x_data = np.array([[1, 2, 0], [5, 4, 3], [1, 2, -1], [3, 1, 0], [2, 4, 2],
                   [4, 1, 2], [-1, 3, 2], [4, 3, 3], [0, 2, 6], [2, 2, 1],
                   [1, -2, -2], [0, 1, 3], [1, 1, 3], [0, 1, 4], [2, 3, 3]])

t_data = np.array([-4, 4, -6, 3, -4,
                   9, -7, 5, 6, 0,
                   4, 3, 5, 5, 1])

print('x_data.shape = ', x_data.shape, ', t_data.shape = ', t_data.shape)
```

```python
# Linear Regression model 생성

model = Sequential() # keras api

# x_data의 shape => (15,3) 이므로 노드를 (3,)으로 설정
model.add(Dense(1, input_shape=(3, ),activation='linear'))
```

```python
# 학습에 사용될 optimizer, 손실 함수 설정

# SGD => 확률적 경사 하강법, mse => 평균 제곱 오차 (caused by linear regression)
learning_rate = 1e-2
model.compile(optimizer=SGD(learning_rate=learning_rate), loss='mse')

print(model.summary()) # 모델 체크
```

```python
from datetime import datetime

start_time = datetime.now()

hist = model.fit(x_data, t_data, epochs=500)

end_time = datetime.now()

print('\nElapsed Time => ', end_time - start_time)
```

```python
print(model.input)
print(model.output)
print(model.weights)

# 예측,  ret_val = 2*x1 -3*x2 + 2*x3

test_data = [ [5, 5, 0], [2, 3, 1], [-1, 0, -1], [10, 5, 2], [4, -1, -2] ]

ret_val = [ 2*data[0] -3*data[1] + 2*data[2]  for data in test_data ]

prediction_val = model.predict(np.array(test_data))

print(prediction_val)
print('==================')
print(ret_val)
```
